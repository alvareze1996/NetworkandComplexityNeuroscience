```python
import ActflowToolbox as actflow
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.io
```

```python
## Import Resting Data
rest=scipy.io.loadmat('RestingTest3.mat')
## Import Task 
task=scipy.io.loadmat('Task.mat')

## Exract data
rest= rest['Resting']
task= task['Task']
```

```python
## ACTFLOW
print("==Activity flow mapping results, correlation-based resting-state FC, 24 task conditions==")
actflowOutput_restFCCorr_bycond = actflow.actflowcomp.actflowtest(task, rest)
```

```python
### Avg Correlation Matrix
AvgRest=scipy.io.loadmat('AverageCorr.mat')
AvgRest=AvgRest['AverageCorr']
fig=actflow.tools.addNetColors_Seaborn(AvgRest)
sns.heatmap(AvgRest)
```

```python
## Predicted Activations
plt.figure(figsize=[7,5])
ax = sns.heatmap(np.mean(actflowOutput_restFCCorr_bycond['actPredVector_bytask_bysubj'],axis=2),center=0,cmap='seismic',cbar=True,yticklabels=100,xticklabels=['Lottery amount', 'Risk'])
ax.figure.suptitle('Predicted activations, corr FC actflow')
ax.set(ylabel='Regions')

### Average Task
AvgTask=scipy.io.loadmat('AverageTask.mat')
AvgTask=AvgTask['AverageTask']

plt.figure(figsize=[7,5])
ax = sns.heatmap(AvgTask,center=0,cmap='seismic',cbar=True,yticklabels=100,xticklabels=['Lottery amount', 'Risk'])
ax.figure.suptitle('Actual activations')
ax.set(ylabel='Regions')
```
